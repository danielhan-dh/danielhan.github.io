```python
data = pd.read_csv('college_data.csv')
```

```python
f, axes = plt.subplots(2,2, figsize = [14,10], gridspec_kw = {'wspace':0.15, 'hspace':0.3})
axes = axes.flatten()
xVars = ['num_students', 'student_staff_ratio', 'international_students', 'female']
xLabels = ['Total Number of Students', 'Student to Staff Ratio', 'Percent of International Students', 'Percent of Female Students']
yVars = 'total_score'
yLabels = 'Total Score'
for i, ax in enumerate(axes):
    hb = ax.hexbin(data = data, x = data[xVars[i]], y = yVars, cmap = 'Blues',
                mincnt = 1, gridsize=12, vmin = 1, vmax = 70)
    axes[i].set_xlabel(xLabels[i], fontsize = 14, labelpad = 8)
    axes[i].set_ylabel(yLabels, fontsize = 14, labelpad = 5)
    axes[i].tick_params(labelsize = 10)
    axes[i].set_facecolor('gainsboro')
    [ax.spines[i].set_visible(False) for i in ax.spines]
colorBar = f.add_axes([0.92, 0.15, 0.02, 0.7])
cb = plt.colorbar(hb, colorBar)
cb.set_label('Frequency', fontsize = 15)

plt.savefig('hex.png', bbox_inches = 'tight', facecolor = 'white') #leave off face_color for transparency
plt.show()
```
